**Loan Elegebility Model**
This project is divided into two parts:
1. Creating a model and trianing a model on dataset
2. Using the model to predict the customer is eligible for loan or not
The model used in this project is **Logistic Regression**

**Specifics:**
   1. Target variable: Loan_Approved
   2. Input variables: Refer to dat dictionary below
   3. Success Criteria: Accury of 70% and above
   4. Mean acurray of the model is 96%
   5. Cibil score has the heighest weightage in the model

**Data Dictionary:**
1. **loan_id**                     Applicant Id
2. **no_of_dependents**            Number of dependents the applicant has
3. **education**                   Highest level of education
4. **self_employed**               Whether self-employed Yes/No
5. **income_annum**                Applicant Annunal Income
6. **loan_amount**                 Loan amount requested
7. **loan_term**                   Term of the Loan in months
8. **cibil_score**                 Cibil score of the applicant
9. **residential_assets_value**    How much the applicant real estate is worth
10. **commercial_assets_value**    The applicant business or income- genrating properties that applicant's owns.
11. **luxury_assets_value**        The total value of non-essential assets the applicant's owns
12. **bank_asset_value**           The total amopunt of money held in the applicant's bank accounts.
13. **loan_status**                 Wheter the loan was approved or not

```python
import pandas as pd
import joblib

# Load the updated model
model = joblib.load('loan_approval_model_v4.pkl')

expected_columns = [
    'no_of_dependents',
    'income_annum',
    'loan_amount',
    'loan_term',
    'cibil_score',
    'residential_assets_value',
    'commercial_assets_value',
    'bank_asset_value',
    'education_ Graduate',
    'self_employed_yes'
]

def check_loan_eligibility():
    print("Enter customer loan application details:\n")

    # --- Inputs with try/except ---

    while True:
        try:
            no_of_dependents = int(input("Number of Dependents: "))
            break
        except error:
            print("Please enter a valid integer.")

    while True:
        education = input("Education (Graduate/Not Graduate): ").strip().lower()
        if education in ['graduate', 'not graduate']:
            break
        print("Please enter 'Graduate' or 'Not Graduate'.")

    while True:
        self_employed = input("Self Employed? (Yes/No): ").strip().lower()
        if self_employed in ['yes', 'no']:
            break
        print("Please enter 'Yes' or 'No'.")

    while True:
        try:
            income_annum = float(input("Annual Income: "))
            break
        except error:
            print("Please enter a valid number.")

    while True:
        try:
            loan_amount = float(input("Loan Amount: "))
            break
        except error:
            print("Please enter a valid number.")

    while True:
        try:
            loan_term = float(input("Loan Term (in months): "))
            break
        except error:
            print("Please enter a valid number.")

    while True:
        try:
            cibil_score = float(input("CIBIL Score: "))
            break
        except error:
            print("Please enter a valid number.")

    while True:
        try:
            residential_assets_value = float(input("Residential Assets Value: "))
            break
        except error:
            print("Please enter a valid number.")

    while True:
        try:
            commercial_assets_value = float(input("Commercial Assets Value: "))
            break
        except error:
            print("Please enter a valid number.")

    while True:
        try:
            bank_asset_value = float(input("Bank Asset Value: "))
            break
        except error:
            print("Please enter a valid number.")

    # --- Prepare input dict ---
    input_data = {
        'no_of_dependents': no_of_dependents,
        'income_annum': income_annum,
        'loan_amount': loan_amount,
        'loan_term': loan_term,
        'cibil_score': cibil_score,
        'residential_assets_value': residential_assets_value,
        'commercial_assets_value': commercial_assets_value,
        'bank_asset_value': bank_asset_value,
        'education_ Graduate': 1 if education == 'graduate' else 0,
        'self_employed_yes': 1 if self_employed == 'yes' else 0
    }

    customer_df = pd.DataFrame([input_data])
    customer_df = customer_df[expected_columns]

    # --- Prediction ---
    prediction = model.predict(customer_df)[0]
    probability = model.predict_proba(customer_df)[0][1]

    print("\nPrediction Result:")
    if prediction == 1:
        print("Loan Approved")
    else:
        print("Loan Not Approved")
    print(f"Approval Probability: {probability:.2f}")

# Run
if __name__ == "__main__":
    check_loan_eligibility()
```
